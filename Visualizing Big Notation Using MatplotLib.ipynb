```python
import numpy as np
import matplotlib.pyplot as plt
from math import log

```

```python
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
a = np.linspace(1,10,1000)

labels = ['Constant','Logarithmic','Linear','Log Linear','Quadratic','Cubic','Exponential' ]
```

```python
notations = [np.ones(a.shape),np.log(a),a,a*np.log(a),a**2,a**3,2**a]

```

```python
plt.figure(figsize = (10,10))
plt.ylim(0,80)

for i in range (len(notations)):
    plt.plot(a,notations[i], label = labels[i])

    
plt.legend(loc=0)
plt.ylabel('Runtime')
plt.xlabel('n')
```
